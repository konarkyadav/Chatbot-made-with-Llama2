## LLaMA2 7B Chatbot downloading and training


````python
# Importing relevant libraries
import torch
import transformers
from transformers import AutoTokenizer, AutoModelForCausalLM
````

````python
# # Import the tokenizer for the "meta-llama/Llama-2-7b-chat-hf" model using Hugging Face's AutoTokenizer.
# This tokenizer is specific to the Llama-2-7b-chat-hf model and is necessary for preprocessing text for the model.
# The 'use_auth_token=True' parameter ensures that the request is authenticated with your Hugging Face token.

tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                          use_auth_token=True)

# Import the actual LLM model from Hugging Face using AutoModelForCausalLM, which is used for language generation tasks.
# 'device_map='auto'' automatically assigns the model to the best available device (CPU or GPU).
# 'torch_dtype=torch.float16' sets the data type for PyTorch tensors to 16-bit floating point, which can reduce memory usage.
# 'use_auth_token=True' is used for authentication with your Hugging Face token.
# The commented lines 'load_in_8bit=True' and 'load_in_4bit=True' are options for further reducing memory usage by
# loading the model in 8-bit or 4-bit precision, but they are not active in this code snippet.

model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                             device_map='auto',
                                             torch_dtype=torch.float16,
                                             use_auth_token=True
                                            #  load_in_8bit=True,
                                            #  load_in_4bit=True
                                             )
````

````python
# Use a pipeline for later
from transformers import pipeline

# # Create a text generation pipeline using the previously loaded model and tokenizer.
# 'pipeline("text-generation")' sets up a pipeline for generating text.
# 'model=model' specifies the model to use for text generation.
# 'tokenizer=tokenizer' sets the tokenizer for pre-processing the input text and post-processing the output.
# 'torch_dtype=torch.bfloat16' uses 16-bit Brain Floating Point format for tensors, optimizing memory and computational efficiency.
# 'device_map="auto"' ensures the pipeline automatically uses the best available device (CPU or GPU).
# 'max_new_tokens=512' limits the maximum number of new tokens to be generated.
# 'do_sample=True' enables probabilistic sampling for diverse and realistic text generation.
# 'top_k=30' limits the sampling pool to the top 30 tokens at each step, balancing creativity and coherence.
# 'num_return_sequences=1' specifies that each prompt should return one sequence of text.
# 'eos_token_id=tokenizer.eos_token_id' sets the end-of-sequence token ID to signal the end of text generation.

pipe = pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 512,
                do_sample=True,
                top_k=30,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )

'''This code snippet sets up a text-generation pipeline using the Hugging Face Transformers library.
It configures the pipeline with the specific model and tokenizer, optimizes for performance with the torch_dtype and device_map parameters,
and tailors the text generation behavior with parameters like max_new_tokens, do_sample, and top_k. This setup is ideal for generating text responses in
various applications, including chatbots and automated content creation.
'''
````

````python
'''This code includes several functions and settings designed for generating and handling responses in a text-based AI model.
It sets up a default system prompt with ethical guidelines, formats prompts, handles text output, and generates responses using the model and tokenizer.
The generate function is key for creating model responses, while parse_text is used for displaying the output in a readable format.
The commented out return assistant_text suggests an option for returning the processed text directly.
'''


import json
import textwrap

# Define special markers for instructions and system messages.

B_INST, E_INST = "[INST]", "[/INST]"
B_SYS, E_SYS = "<<SYS>>\n", "\n<</SYS>>\n\n"

# Set a default system prompt that instructs the assistant to be helpful, respectful, honest, and safe.
# It emphasizes the importance of avoiding harmful, unethical, or biased content.

DEFAULT_SYSTEM_PROMPT = """\
You are a helpful, respectful and honest assistant. Always answer as helpfully as possible, while being safe. Your answers should not include any harmful, unethical, racist, sexist, toxic, dangerous, or illegal content. Please ensure that your responses are socially unbiased and positive in nature.

If a question does not make any sense, or is not factually coherent, explain why instead of answering something not correct. If you don't know the answer to a question, please don't share false information."""

# Function to generate a formatted prompt with the given instruction and an optional system prompt.

def get_prompt(instruction, new_system_prompt=DEFAULT_SYSTEM_PROMPT ):
    SYSTEM_PROMPT = B_SYS + new_system_prompt + E_SYS
    prompt_template =  B_INST + SYSTEM_PROMPT + instruction + E_INST
    return prompt_template

# Function to cut off the text at the appearance of a specified prompt, ensuring clean output.

def cut_off_text(text, prompt):
    cutoff_phrase = prompt
    index = text.find(cutoff_phrase)
    if index != -1:
        return text[:index]
    else:
        return text

# Function to remove a specified substring from a string.

def remove_substring(string, substring):
    return string.replace(substring, "")


# Function to generate a response based on the input text using the model and tokenizer.

def generate(text):
    prompt = get_prompt(text)
    with torch.autocast('cuda', dtype=torch.bfloat16):
        inputs = tokenizer(prompt, return_tensors="pt").to('cuda')
        outputs = model.generate(**inputs,
                                 max_new_tokens=512,
                                 eos_token_id=tokenizer.eos_token_id,
                                 pad_token_id=tokenizer.eos_token_id,
                                 )
        final_outputs = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]
        final_outputs = cut_off_text(final_outputs, '</s>')
        final_outputs = remove_substring(final_outputs, prompt)

    return final_outputs#, outputs

# Function to format and print the text with a specified width for better readability.

def parse_text(text):
        wrapped_text = textwrap.fill(text, width=100)
        print(wrapped_text +'\n\n')
        # return assistant_text
````

## LangChain basics

````python
from langchain import HuggingFacePipeline
from langchain import PromptTemplate,  LLMChain
````

````python
'''This code snippet is setting up an instance of HuggingFacePipeline, a class presumably designed to interact with Hugging Face's models and pipelines.
The pipe object, created earlier, is passed as the pipeline configuration. The model_kwargs with 'temperature':0 alters the behavior of the model, making its
responses more predictable and less varied. This setup is typically used in scenarios where consistent and less random outputs are preferred.'''

# Initialize an instance of the HuggingFacePipeline.
# The 'pipeline' parameter is set to the 'pipe' object previously defined, which contains the configuration
# for the text generation pipeline, including the model, tokenizer, and other settings.
# The 'model_kwargs' parameter is a dictionary that sets additional keyword arguments for the model.
# Here, 'temperature' is set to 0, which influences the randomness of the output.
# A lower temperature results in less random and more deterministic outputs from the model.

llm = HuggingFacePipeline(pipeline = pipe, model_kwargs = {'temperature':0})
````

````python
'''This code snippet sets up a translation task using a pre-defined LLMChain class, which likely serves as a wrapper or interface for language model operations.
The PromptTemplate and LLMChain classes are used here to structure the input (in this case, text for translation) and to define the behavior of the
language model (in this case, for translation purposes). The template variable combines the instruction for translation with the system prompt, ensuring the AI
understands its role as a translation assistant. The prompt and llm_chain objects are then used to operationalize this setup.'''


# Define a system prompt that positions the AI as an advanced assistant specialized in translation.
system_prompt = "You are an advanced assistant that excels at translation. "

# Define an instruction for the AI to translate a given text from English to French.
instruction = "Convert the following text from English to French:\n\n {text}"

# Use the 'get_prompt' function to create a formatted template combining the system prompt and instruction.
# This formatted template will guide the AI in understanding and executing the translation task.
template = get_prompt(instruction, system_prompt)
print(template)

# Import or define the PromptTemplate class.
# Create an instance of PromptTemplate with the formatted template and define 'text' as an input variable.
# This sets up a structured format for the input text that needs to be translated.
prompt = PromptTemplate(template=template, input_variables=["text"])

# Import or define the LLMChain class.
# Initialize an instance of LLMChain with the prompt configuration and the language model pipeline 'llm'.
# LLMChain is presumably a class that facilitates interaction with the language model for specific tasks.
llm_chain = LLMChain(prompt=prompt, llm=llm)
````

````python
text = "how are you today?"
output = llm_chain.run(text)

parse_text(output)
````

## Simple Chatbot

````python
from langchain.memory import ConversationBufferMemory
from langchain import LLMChain, PromptTemplate
````

````python
instruction = "Chat History:\n\n{chat_history} \n\nUser: {user_input}"
system_prompt = "You are a helpful assistant, you always only answer for the assistant then you stop. read the chat history to get context"

template = get_prompt(instruction, system_prompt)
print(template)
````

````python
'''This code snippet is concerned with setting up a structured prompt for conversational AI and managing the conversation history.
The PromptTemplate class is configured to include both the chat history and the user's current input, ensuring that the AI model can generate contextually
relevant responses. The ConversationBufferMemory class is used to handle the conversation history, keeping track of the dialogue for better continuity and
understanding of the conversation's context.'''

# Create an instance of PromptTemplate, specifying the template and the input variables 'chat_history' and 'user_input'.
# This sets up a structured prompt format that integrates both the chat history and the current user input,
# which is essential for context-aware responses in conversational AI applications.

prompt = PromptTemplate(
    input_variables=["chat_history", "user_input"], template=template
)
# Initialize an instance of ConversationBufferMemory.
# This class is designed to store and manage the chat history, using 'chat_history' as the memory key.
# The chat history is critical for maintaining context and coherence in ongoing conversations.
memory = ConversationBufferMemory(memory_key="chat_history")
````

````python
'''In this code snippet, an LLMChain instance is being created, integrating several key components for a conversational AI system.
The llm parameter connects it to the pre-configured language model, while prompt sets up the structured interaction format. The verbose parameter enables
detailed output, useful for monitoring or debugging, and memory incorporates the conversation history, crucial for maintaining context in ongoing interactions.
This setup provides a comprehensive framework for managing AI-driven conversations.'''


# Initialize an instance of LLMChain.
# This class facilitates the interaction between the language model and the structured prompt.
# 'llm' is set to the previously defined language model pipeline.
# 'prompt' is the PromptTemplate instance set up with the necessary input variables and template.
# 'verbose=True' enables detailed logging or output, which is useful for debugging or understanding the model's processing.
# 'memory' is set to the ConversationBufferMemory instance, enabling the LLMChain to utilize and update the chat history.
# This configuration is essential for conversational AI applications, ensuring contextual awareness and continuity.
llm_chain = LLMChain(
    llm=llm,
    prompt=prompt,
    verbose=True,
    memory=memory,
)
````

````python
'''This line of code is executing the predict method of the llm_chain object, which is set up to handle language model predictions in a conversational context.
The method takes the user input "Hi, my name is Konark" and processes it to generate an appropriate response, considering both the context established by the
conversation history and the guidelines set in the prompt template. This functionality is key in interactive chatbot applications.'''

# Call the 'predict' method on the 'llm_chain' object, passing in a user input.
# In this case, the user input is "Hi, my name is Konark".
# The 'predict' method uses the LLMChain instance to process this input, considering the structured prompt and memory.
# It generates a response based on the language model's understanding and the conversation context provided by the memory.
# This method is typically used in conversational AI to generate responses to user inputs in real-time.

llm_chain.predict(user_input="Hi, my name is Konark")
````
